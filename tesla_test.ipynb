```python
#import libraries
import pandas as pd
from pathlib import Path
```

```python
#import file
csv_path = Path("./v_tsla_newsapi.csv")
csv_path_price= Path("./TSLA.csv")
#tesla=pd.read_csv(csv_path, parse_dates=True, infer_datetime_format=True) datetime format wrong
tesla=pd.read_csv(csv_path, parse_dates=['date'], infer_datetime_format=True)
stock_price=pd.read_csv(csv_path_price, parse_dates=['Date'])#, index_col="Date", infer_datetime_format=True)
```

```python
#inspect csv files and drop unnecessary columns
stock_price=stock_price.drop(columns=["Open","High","Low","Adj Close","Volume"])
```

```python
tesla=tesla.groupby(['Date']).mean()*100   #pct_change()*100 #or use .mean()
```

```python
new_tesla=tesla.sort_values("Date", ascending = False)
new_tesla=new_tesla.reset_index()
```

```python
#new_tesla.head(),
new_stock_price=new_stock_price.reset_index()
```

```python
#new_stock_price
new_stock_price=new_stock_price.drop(columns=['index'])
```

```python
#change new stock data and see if it plots better
new_stock_price['Close']=new_stock_price['Close'].pct_change()*100
```
